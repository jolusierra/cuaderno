# Empezamos con los gráficos




A continuación vamos a crear una lista de 100 números:


```python
import numpy as np
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
X=np.linspace(0,20,100)
```

```python
X
```

Ahora generamos una nueva lista Y al aplicar la función
$$ x^2 -3x +5$$

```python
Y=X**2-3*X+5
```

```python
Y
```

Representamos gráficamente Y frente a X

```python
fig,ax=plt.subplots()
ax.plot(X,Y)
```

Añadimos título a la gráfica, nombre a los ejes y un texto explicativo dentro de la gráfica:

```python
fig,ax=plt.subplots()
ax.plot(X,Y)
ax.set_title('Gráfica de Y frente a X')
ax.set_xlabel('eje X')
ax.set_ylabel('eje Y')
ax.text(15,150,r'$y=x^2-3x+5$')
plt.show()
```

También se puede representar en una misma gráfica distintas funciones con sus correspondientes leyendas.
La lista "t" contiene 20 valores comprendidos entre 0 y 5, y se corresponde con distintos instantes de tiempo.
Se representarán las gráficas de posición frente a tiempo para los movimientos descritos por las siguientes ecuaciones:

$$ x=2t $$
$$ x=t^2 $$

```python
t=np.linspace(0,5,20)
fig,ax=plt.subplots()
ax.plot(t,2*t,label='MRU')
ax.plot(t,t**2,label='MRUA')
ax.set_title('Gráfica de posición frente a tiempo')
ax.set_xlabel('tiempo (s)')
ax.set_ylabel('posición (m)')
ax.legend()
ax.text(4,15,r'$x=t^2$')
ax.text(4,7,r'$x=2t$')
fig.savefig("figura.pdf")
```

```python
from sympy import *
```

Definimos las variables "x" e "y". Posteriormente calculamos la derivada de sen(x) y la integral indefinida de $ 3x^2 $

```python
x,y=symbols('x y')
```

```python
diff(sin(x),x)
```

```python
integrate(3*x**2,x)
```

Podemos evaluar el valor numérico de distintas constantes con un determinado número de cifras decimales:

```python
pi
```

Se pueden simplificar expresiones algebraicas. Por ejemplo: $ (x-2)^2-x^2 $

```python
(x-2)**2-x**2
```

```python
((x-2)**2-x**2).expand()
```

Se puede generar código en latex para una determinada expresión matemática.

```python
latex((x-2)**2-x**2)
```

La expresión algebraica anterior es:

$$- x^{2} + \left(x - 2\right)^{2}$$

Asimismo, se pueden calcular integrales definidas y límites.

Por ejemplo: $ \int_0^{10} 2x \; dx $

```python
integrate(2*x,(x,0,10))
```

```python
diff(2*x,x)
```

```python
limit((2*x+2*y-2*x)/y,y,0)
```

```python
integ=Integral(3*x**3-x**2+7,(x,0,10))
```

```python
integ.evalf()
```

```python
latex(integ)
```

La integral anterior es:

$$\int_{0}^{10} \left(3 x^{3} - x^{2} + 7\right)\, dx$$


A continuación resolveremos la ecuación:

$$ 2x-6=0 $$

```python
solve(2*x-6,x)
```

Ahora resolveremos la ecuación de 2º grado en "x", quedando la variable "y" como un parámetro:

$$2x^2-3x+y=0 $$

```python
solve(2*x**2-3*x+y,x)
```

La soluciones son:

$$x=\frac{3 \pm \sqrt{9-8y}}{4} $$

También se puede resolver un sistema de ecuaciones como:

$$2x-y=0 $$
$$3x+4y=0 $$

```python
solve([2*x-y,3*x+4*y],[x,y])
```

Definimos la función f(x)

$$ f(x)=x^2-4$$

```python
def f(x):
  return x**2-4
```

```python
f(2)
```

```python
f(-1)
```

Derivamos la función f(x)

$$ \frac{df}{dx} $$

```python
diff(f(x),x)
```

Calculamos la integral indefinida de f(x):

$$ ∫ f(x) \; dx$$

```python
integrate(f(x),x)
```

Asimismo, se puede calcular la integral definida en un determinado intervalo. Por ejemplo:

$$ \int_0^{10} f(x) \; dx  $$

```python
integrate(f(x),(x,0,10))
```
